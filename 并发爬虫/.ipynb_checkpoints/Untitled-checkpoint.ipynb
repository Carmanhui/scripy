```python
import os
import re
import time
from urllib import request
from concurrent.futures import ProcessPoolExecutor
from bs4 import BeautifulSoup
WWW_BIQUGE_CM = 'https://www.piaohua.com'
```

```python
def __fetch_html(url,decode='UTF-8'):
    req = request.Request(url)
    req.add_header('User-Agent','Mozilla/5.0')
    with request.urlopen(req) as res:
        return res.read().decode(decode)
    
def __parse_title_and_hrefs(index):
    html = __fetch_html(f'{WWW_BIQUGE_CM}/{index}')
#     print(html)
    soup = BeautifulSoup(html,'html.parser')
    links = soup.find_all('a',href = re.compile(rf'html/kehuan/2018'))
    hrefs = list(map(lambda x:x['href'],links))
    
#     print(hrefs)
    return hrefs
# #     print(hrefs[10:])
#     title = soup.h1.string
# #     print(f'title:{title}\nhrefs:size = {len(hrefs)}')
#     return title,hrefs[10:]
index = 'html/kehuan/index.html'
hrefs = __parse_title_and_hrefs(index)

for href in hrefs:
    html = __fetch_html(f'{WWW_BIQUGE_CM}{href}')
#     print(html)
    soup = BeautifulSoup(html,'html.parser')
    title = soup.h1.string
    print(title)
    links = soup.find_all(href=re.compile(rf'urn:btih'))
    print(links)
```
